```python
import numpy as np
import matplotlib.pyplot as plt
import nibabel as nib
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim
from skimage.metrics import mean_squared_error as mse
```

```python
img_3d_t2 = np.array(nib.load("Day2/testimg_3d_t2.nii").get_fdata().astype(np.float32))
```

```python
img_3d_t2.shape
```

```python
import numpy as np

def NormalizeData(data):
    return (data - np.min(data)) / (np.max(data) - np.min(data))


def range(data):
    return (data.min(), data.max())

img1 = img_3d_t2[:,:,100]
img2 = img_3d_t2[:,:,101]
img1n = NormalizeData(img1)
img2n = NormalizeData(img2)
```

```python
# current
print(range(img1))
print(range(img2))
```

```python
#  after normalize
print(range(img1n))
print(range(img2n))
```

```python
plt.imshow(np.hstack((img1,img2)))
```

```python
print(psnr(img1n, img2n))
print(ssim(img1n, img2n))
print(mse(img1n, img2n))
plt.imshow(np.hstack((img1,img2)))
```

```python
img3 = img_3d_t2[:,:,100]
img4 = img_3d_t2[:,:,110]
img3n = NormalizeData(img3)
img4n = NormalizeData(img4)
print(psnr(img3n, img4n))
print(ssim(img3n, img4n))
print(mse(img3n, img4n))
```
